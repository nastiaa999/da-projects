## Задача 2. Анализ рынка акций

1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv). Сохраните DataFrame в `DF_COMPANIES`

```python
all_data = []
folder = 'data/stock'

for file in os.listdir(folder):
    if file.endswith('.csv'):
        stock_name = file.replace('.csv', '')
        df = pd.read_csv(f'{folder}/{file}')
        if 'Close' not in df.columns:
            continue
        df = df[['Date', 'Close']].copy()
        df['Date'] = pd.to_datetime(df['Date'])
        df = df.set_index('Date') #установим индексы
        df = df.rename(columns={'Close': stock_name}) 
        all_data.append(df)

#объединяем всё в один df 
DF_COMPANIES = pd.concat(all_data, axis=1)
DF_COMPANIES = DF_COMPANIES.sort_index(axis=1)
DF_COMPANIES
```

2. Рассчитайте корреляционную матрицу для всех акций и сохраните её в  `DF_CORR`

```python
DF_CORR = DF_COMPANIES.corr()

#проверим хэш
import hashlib
def robust_hash(obj):
    if isinstance(obj, (pd.Series, pd.DataFrame)):
        obj = obj.values
    if isinstance(obj, np.ndarray):
        obj = obj.flatten().tolist()
    if isinstance(obj, list):
        processed = []
        for item in obj:
            if isinstance(item, (float, np.floating)):
                if np.isnan(item):
                    processed.append("nan")
                else:
                    processed.append(f"{item:.4f}")
            else:
                processed.append(str(item))
        string = str(processed)
    else:
        string = str(obj)
    return hashlib.sha256(string.encode("utf-8")).hexdigest()[:8]
# Считаем твой хеш
my_hash = robust_hash(DF_CORR.values.flatten())
print( my_hash)
DF_CORR
```

3. Отобразите корреляционную матрицу в виде диаграммы.

```python
plt.figure(figsize=(12, 10))
sns.heatmap(DF_CORR, 
            annot=True,          # показываем числа в ячейках
            cmap='RdYlGn',       # красивая цветовая схема: красный — отрицательная, зелёный — положительная
            center=0,            # 0 в центре (чтобы было симметрично)
            fmt='.2f',           # 2 знака после запятой
            linewidths=0.5,      # тонкие линии между ячейками
            cbar_kws={'label': 'Корреляция'},  # подпись шкалы
            square=True)         # квадратные ячейки
plt.title('Корреляционная матрица акций', fontsize=16, pad=20)
plt.xlabel('Акции')
plt.ylabel('Акции')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

4. В соответствии с Вашим вариантом (Вариант 2. Netflix (NFLX)) определите:

    1) акцию с максимальной положительной корреляцией (max)
    2) акцию с максимальной отрицательной корреляцией (min)
    3) акцию с минимальной корреляцией (которая больше всего соответствует отсутствию какой-либо корреляции (none))

Сохраните результаты в списке `LST_COMP_CORR` в указанном порядке

```python
correlations = DF_CORR['NFLX'].drop('NFLX')
max_pos_corr = correlations.idxmax()
max_neg_corr = correlations.idxmin()
none_corr = (correlations - 0).abs().idxmin()
LST_COMP_CORR = [max_pos_corr, max_neg_corr, none_corr]
LST_COMP_CORR
```

5. Постройте диаграммы разброса (Ваша компания - Компания с min), (Ваша компания - Компания с max), (Ваша компания - Компания с none). По x расположите цену вашей компании, по y - цену компании в паре

```python
base_stock = 'NFLX'
max_stock  = LST_COMP_CORR[0]
min_stock  = LST_COMP_CORR[1]  
none_stock = LST_COMP_CORR[2]   

x = DF_COMPANIES[base_stock]
y_max  = DF_COMPANIES[max_stock]
y_min  = DF_COMPANIES[min_stock]
y_none = DF_COMPANIES[none_stock]

fig, axes = plt.subplots(1, 3, figsize=(16, 6))

#макс. положительная корреляция
axes[0].scatter(x, y_max, alpha=0.6, color='green', s=30)
axes[0].set_title(f'NFLX и {max_stock} (max положительная)\nкорр = {DF_CORR.loc[base_stock, max_stock]:.3f}', 
                  fontsize=12, pad=15)
axes[0].set_xlabel('Netflix (NFLX)')
axes[0].set_ylabel(max_stock)

#макс. отрицательная корреляция
axes[1].scatter(x, y_min, alpha=0.6, color='red', s=30)
axes[1].set_title(f'NFLX и {min_stock} (max отрицательная)\nкорр = {DF_CORR.loc[base_stock, min_stock]:.3f}', 
                  fontsize=12, pad=15)
axes[1].set_xlabel('Netflix (NFLX)')
axes[1].set_ylabel(min_stock)

#почти нет корреляции 
axes[2].scatter(x, y_none, alpha=0.6, color='gray', s=30)
axes[2].set_title(f'NFLX и {none_stock} (близко к 0)\nкорр = {DF_CORR.loc[base_stock, none_stock]:.3f}', 
                  fontsize=12, pad=15)
axes[2].set_xlabel('Netflix (NFLX)')
axes[2].set_ylabel(none_stock)

fig.suptitle('Диаграммы разброса: Netflix и другие акции', fontsize=16, y=1.02)

plt.tight_layout()
plt.show()
```

<i> На первой диаграмме компании с максимальной положительной корреляцией: когда акция Netflix растёт, акция тоже почти всегда растёт. Когда Netflix падает, вторая тоже падает

Для второго графика максимальная отрицательная корреляция не близка к -1, точки разбросаны хаотично. Если одна акция растет, вторая может падать и наоборот

На третьем графике также отсутствует какой-либо наклон, цены акций не зависят друг от друга </i>
